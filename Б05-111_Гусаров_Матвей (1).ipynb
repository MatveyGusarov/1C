__Задача 1. (3 балла)__ Пусть функция $f:\mathbb{R} \to \mathbb{R}$ задана следующим образом $f(x) = \max\{-x, x, x^2\}$. Найдите субдифференциал данной функции $\partial f(x)$.

Субдифференциал функции $\( f \)$ в точке $\( x \)$, обозначаемый как $\( \partial f(x) \)$, является обобщением понятия производной. Для функции, которая не является дифференцируемой в классическом смысле (например, из-за изломов или углов), субдифференциал включает все возможные "наклоны" линий, касающихся графика функции в этой точке.

Функция $\( f(x) = \max\{-x, x, x^2\} \)$ представляет собой максимум трех функций: $\(-x\)$, $\(x\)$, и $\(x^2\)$. Для нахождения субдифференциала этой функции, нам нужно рассмотреть поведение каждой из этих функций и определить, какие из них определяют $\( f(x) \)$ в каждой точке.

1. Когда $\( f(x) = -x \)$, это означает, что $\(-x \geq x\) и \(-x \geq x^2\)$. Это возможно, когда $\( x \leq 0 \)$ и $\( x^2 \leq -x \)$, то есть $\( x \in [-1, 0] \)$. В этом случае, производная $\(-x\)$ равна $\(-1\)$.

2. Когда $\( f(x) = x \)$, это означает, что $\( x \geq -x \)$ и $\( x \geq x^2 \)$. Это возможно, когда $\( x \geq 0 \)$ и $\( x^2 \leq x \)$, то есть $\( x \in [0, 1] \)$. В этом случае, производная $\( x \)$ равна $\( 1 \)$.

3. Когда $\( f(x) = x^2 \)$, это означает, что $\( x^2 \geq x \) и \( x^2 \geq -x \)$. Это возможно, когда $\( x \leq -1 \)$ или $\( x \geq 1 \)$. В этом случае, производная $\( x^2 \)$ равна $\( 2x \)$.

Таким образом, субдифференциал $\( \partial f(x) \)$ можно определить следующим образом:

- Для $\( x < -1 \)$: $\( \partial f(x) = \{2x\} \)$, так как $\( f(x) = x^2 \)$ и производная $\( 2x \)$.
- Для $\( x = -1 \)$: $\( \partial f(x) = [-2, -1] \)$, так как в этой точке возможны наклоны от обеих функций $\( -x \)$ и $\( x^2 \)$.
- Для $\( -1 < x < 0 \)$: $\( \partial f(x) = \{-1\} \)$, так как $\( f(x) = -x \)$ и производная $\(-1\)$.
- Для $\( x = 0 \)$: $\( \partial f(x) = [-1, 1] \)$, так как в этой точке возможны наклоны от всех трех функций.
- Для $\( 0 < x < 1 \)$: $\( \partial f(x) = \{1\} \)$, так как $\( f(x) = x \)$ и производная $\(1\)$.
- Для $\( x = 1 \)$: $\( \partial f(x) = [1, 2] \)$, так как в этой точке возможны наклоны от обеих функций $\( x \)$ и $\( x^2 \)$.
- Для $\( x > 1 \)$: $\( \partial f(x) = \{2x\} \)$, так как $\( f(x) = x^2 \)$ и производная $\( 2x \)$.

__Задача 3. (5 баллов)__ Пусть функция $f: \mathbb{R}^n \to \mathbb{R}$ задана следующим образом $f(x) = \exp\left(\|Ax - b\|_p\right)$, где $A \in \mathbb{R}^{m \times n}$, $b \in \mathbb{R}^m$, $p \in [1; +\infty]$. Найти субдифференциал $\partial f(x)$.

Чтобы найти субдифференциал функции $f(x)=\exp \left(\|A x-b\|_p\right)$, нам нужно использовать правила дифферешцировапия композиции фушкций и пормы.

Сначала рассмотрим $g(x)=\|A x-b\|_p$. Субдифференциал нормы $\|\cdot\|_p$ зависит от значения $p$ :
1. Для $1<p<\infty, p$-норма дифферешируема везде, кроме точки 0 , и её субдифференциал в точке $z$ (где $z=A x-b$ ) задается как
$$
\partial\|z\|_p=\left\{\frac{z}{\|z\|_p^{p-1}}\right\}
$$

для $z \neq 0$. Для $z=0$, субдифференциал является замкнутым шаром радиуса 1 в $\ell_q$-норме, где $\frac{1}{p}+\frac{1}{q}=1$.
2. Для $p=1$, субдифферешиал пормы $\|z\|_1$ в точке $z$ является замкнутым выпуклым множеством, элементы которого имеют $\infty$-норму, не превыпающую 1. То есть, это множество векторов, каждый из которых в каждой коордишате пришимает зшачение из отрезка $[-1,1]$, соответствующее заку соответствующей координаты вектора $z$.
3. Для $p=\infty$, субдифференциал нормы $\|z\|_{\infty}$ в точке $z$ состоит из векторов, каждая координата которых лежит в отрезке $[-1,1]$, если соответствующая коордишата вектора $z$ равна $\|z\|_{\infty}$, и равна 0 в противном случае.

Затем, рассмотрим функцию $f(x)$. Её субдифференциал можно найти, используя правило цепочки для субдифференциалов. Пусть $h(x)=\exp (x)$. Тогда субдифферешиал $f(x)$ будет определяться как
$$
\partial f(x)=\exp \left(\|A x-b\|_p\right) \cdot \partial g(x)
$$

где $\partial g(x)$ - субдифференциал $g(x)$ в точке $x$, умноженный на матрицу $A$.
Поскольку $\partial g(x)$ зависит от зачешия $p$ и кошкретшы зачеший $A x-b$, для полного решения необходимо уточнить эти значения. Но общая форма субдифференциала $f(x)$ будет иметь вид
$$
\partial f(x)=\exp \left(\|A x-b\|_p\right) \cdot A^T \cdot \partial\|A x-b\|_p
$$

где $A^T$ - транспонированная матрица $A$, а $\partial\|A x-b\|_p$ - субдифференциал нормы $\|A x-b\|_p$, определенный выше для различных значений $p$.

__а). (5 баллов)__ Пусть $\lambda_{\max}: \mathbb{S}^d \to \mathbb{R}$ - функция максимального собственного значения, заданная на $\mathbb{S}^d$. Найдите субдифференциал $\partial \lambda_{\max}(X)$. Здесь $\mathbb{S}^d$ - симметричные матрицы.

_Hint:_ воспользуйтесь вариационным представлением $\lambda_{\max}$ и формулой для субдифференциала максимума.

Функция максимального собственного значения $\lambda_{\max }: \mathbb{S}^d \rightarrow \mathbb{R}$ для симметричной матрицы $X \in \mathbb{S}^d$ может быть представлена с помощью вариационного представления:
$$
\lambda_{\max }(X)=\max _{\|v\|=1} v^T X v
$$

где $v$ - единичный вектор в $\mathbb{R}^d$.
Субдифференциал функции максимального собственного значения $\partial \lambda_{\max }(X)$ состоит из градиентов этой функции в точках максимума. Иными словами, это множество всех матриц, которые можно получить как внешние произведения собственньх векторов, соответствующих максимальному собственному значению, на себя. Это следует из того, что производная квадратичной формы $v^T X v$ по матрице $X$ равна $v v^T$, где $v$ - направление производной.

Таким образом, если $X$ имеет максимальное собственное значение $\lambda_{\max }(X)$, и $u_1, u_2, \ldots, u_k$ - собственные векторы, соответствующие этому собственному значению, то субдифференциал $\partial \lambda_{\max }(X)$ можно описать следующим образом:
$$
\partial \lambda_{\max }(X)=\left\{\sum_{i=1}^k c_i u_i u_i^T: c_i \geq 0, \sum_{i=1}^k c_i=1\right\}
$$

Это множество всех конвексных комбинаций внешних произведений собственных векторов, соответствующих максимальному собственному значению.

Если максимальное собственное значение является кратным (т.е., у матрицы $X$ есть несколько линейно независимых собственньх векторов, соответствующих этому значению), то субдифференциал включает внешние произведения всех этих собственньх векторов. Если же максимальное собственное значение простое (уникальное), то субдифференциал состоит из одной матрицы $u u^T$, где $u$ - единственный собственный вектор, соответствующий максимальному собственному значению.

Функция максимального собственного значения $\lambda_{\max }(X)$ для симметричной матрицы $X$ будет дифференцируема в точке $X$, если и только если максимальное собственное значение матрицы $X$ является простым. Это утверждение обусловлено свойствами гладкости функции в точках, где её максимум достигается на уникальном направлении.

### Доказательство
1. Если максимальное собственное значение простое (единственное)

Пусть $\lambda_{\max }(X)$ - максимальное собственное значение матрицы $X$, и пусть $u$ - соответствующий нормированный собственный вектор. Так как $\lambda_{\max }(X)$ простое, $u$ является единственным собственным вектором, соответствующим этому собственному значению (с точностью до умножения на скаляр).

В этом случае, вариационное представление максимального собственного значения можно записать как
$$

$$

Градиент функции $\lambda_{\max }(X)$ в точке $X$ равен внешнему произведению $u$ на себя, то есть
$$
\nabla \lambda_{\max }(X)=u u^T
$$

Этот результат следует из того факта, что производная квадратичной формы $v^T X v$ по $X$ равна $v v^T$. Таким образом, функция дифференцируема в точке $X$.
2. Если максимальное собственное значение имеет кратность более 1
Если максимальное собственное значение матрицы $X$ имеет кратность более 1 , то существует несколько линейно независимых собственных векторов, соответствующих этому значению. В этом случае, направление, в котором достигается максимум функции $\lambda_{\max }(X)$, не является уникальным, что приводит к тому, что функция не дифференцируема в точке $X$. Это связано с тем, что градиент функции в такой точке не определен однозначно.

Таким образом, функция $\lambda_{\max }(X)$ дифференцируема в точке $X$ тогда и только тогда, когда максимальное собственное значение матрицы $X$ является простым, и её градиент в такой точке равен $u u^T$, где $u$ - собственный вектор, соответствующий максимальному собственному значению.

__Задача 1. (всего 35 баллов)__ Вновь рассмотрим задачу минимизации эмпирического риска:
\begin{equation}
\min_{w \in \mathbb{R}^d} \frac{1}{n} \sum\limits_{i=1}^n \ell (g(w, x_i), y_i) + \frac{\lambda}{2} \| w\|^2_2,
\end{equation}
где $\ell: \mathbb{R} \times \mathbb{R} \to \mathbb{R}$ - функция потерь, $g : \mathbb{R}^d \times \mathbb{R}^x \to \mathbb{R}$ - модель, $w$ - параметры модели, $\{x_i, y_i\}_{i=1}^n$ - выборка данных из векторов признаков $x_i \in \mathbb{R}^x$ и меток $y_i \in \mathbb{R}$.

Используем линейную модель $g(w, x) = w^T x$ и логистическую/сигмоидную функцию потерь: $\ell(z,y) = \ln (1 + \exp(-yz))$ (__Важно: $y$ должен принимать значения $-1$ или $1$__). Как уже знаем, полученная задача называется задачей логистической регрессии.

Данную задачу можно переписать следующим образом:
\begin{equation}
\min_{w \in \mathbb{R}^d} f(w) := \frac{1}{s} \sum\limits_{j=1}^s f_j(w) := \frac{1}{s} \sum\limits_{j=1}^s \left[\frac{1}{b} \sum\limits_{i=1}^b l (g(w, x_{(j-1)b + i}), y_{(j-1)b + i}) + \frac{\lambda}{2} \| w\|^2_2\right],
\end{equation}
где $b$ - размер батча, $s$ - количество батчей, а $b s = n$ - полный размер выборки.

Градиент функции $f_j$:
$$
\nabla f_j(w) = \frac{1}{b} \sum_{i=1}^b \frac{-y_{(j-1)b + i} x_{(j-1)b + i}}{1 + \exp(y_{(j-1)b + i} w^Tx_{(j-1)b + i})}.
$$
Константу Липшица градиента $\nabla f_j$ можно оценить, как $L_j = \frac{1}{4b} \sum\limits_{i=1}^b \| x_{(j-1)b + i} \|^2_2$.

Метод стохастического градиентного спуска (SGD) для данной задачи записывается следующим образом:
$$
w^{k+1} = w^k - \gamma_k \nabla f_{j_k} (w^k),
$$
где номер $j_k$ генерируется независимо и равномерно из $\{1, \ldots, s \}$.

```python
import numpy as np

def logistic_loss_and_gradient(w, X, y, lambda_reg):
    """
    Вычисляет логистические потери и градиент.

    Параметры:
    w (np.ndarray): Вектор весов.
    X (np.ndarray): Матрица признаков.
    y (np.ndarray): Вектор меток.
    lambda_reg (float): Параметр регуляризации.

    Возвращает:
    float: Значение логистических потерь.
    np.ndarray: Градиент потерь.
    """
    # Количество образцов
    n = X.shape[0]

    # Вычисление логистических потерь
    z = y * np.dot(X, w)
    loss = np.mean(np.log(1 + np.exp(-z))) + (lambda_reg / 2) * np.dot(w, w)

    # Вычисление градиента
    gradient = np.mean(
        -y[:, np.newaxis] * X / (1 + np.exp(z[:, np.newaxis])), axis=0
    ) + lambda_reg * w

    return loss, gradient

# Пример использования
# Определите параметр регуляризации
lambda_reg = 0.1

# Инициализация весов случайным образом
w = np.random.randn(X_train.shape[1])

# Вычисление потерь и градиента
loss, grad = logistic_loss_and_gradient(w, X_train, y_train, lambda_reg)

print("Потери:", loss)
print("Градиент:", grad)
```

```python
import numpy as np

def batch_generator(X, y, batch_size):
    """
    Генератор батчей для датасета.

    Параметры:
    X (np.ndarray): Матрица признаков.
    y (np.ndarray): Вектор меток.
    batch_size (int): Размер батча.

    Генерирует:
    np.ndarray, np.ndarray: Батчи признаков и меток.
    """
    n_samples = X.shape[0]
    indices = np.arange(n_samples)
    np.random.shuffle(indices)

    for start_idx in range(0, n_samples, batch_size):
        end_idx = min(start_idx + batch_size, n_samples)
        batch_idx = indices[start_idx:end_idx]

        yield X[batch_idx], y[batch_idx]

def sgd(X, y, batch_size, epochs, learning_rate, lambda_reg):
    """
    Выполнение стохастического градиентного спуска.

    Параметры:
    X (np.ndarray): Матрица признаков.
    y (np.ndarray): Вектор меток.
    batch_size (int): Размер батча.
    epochs (int): Количество эпох.
    learning_rate (float): Скорость обучения.
    lambda_reg (float): Параметр регуляризации.

    Возвращает:
    np.ndarray: Обновленный вектор весов.
    """
    w = np.random.randn(X.shape[1])
    n = X.shape[0]

    for epoch in range(epochs):
        for X_batch, y_batch in batch_generator(X, y, batch_size):
            loss, grad = logistic_loss_and_gradient(w, X_batch, y_batch, lambda_reg)
            w -= learning_rate * grad

    return w

# Пример использования
# Параметры SGD
batch_size = 64
epochs = 10
learning_rate = 0.01
lambda_reg = 0.1

# Выполнение SGD
w_sgd = sgd(X_train, y_train, batch_size, epochs, learning_rate, lambda_reg)
```

Для более честного сравнения можно использовать следующий подход:

- **Нормализация по количеству обработанных примеров**: Вместо того, чтобы смотреть на номер итерации, можно отслеживать количество обработанных примеров. Это даст каждому батчу равное "количество шансов" для обновления весов.

```python
# Определение функции потерь и её градиента
def logistic_loss_and_gradient(w, X_batch, y_batch, lambda_reg):
    z = np.dot(X_batch, w)
    loss = np.mean(np.log(1 + np.exp(-y_batch * z))) + (lambda_reg / 2) * np.sum(w * w)
    gradient = -(X_batch.T @ (y_batch / (1 + np.exp(y_batch * z)))) / len(y_batch) + lambda_reg * w
    return loss, gradient

# Функция для разделения данных на батчи
def batch_generator(X, y, batch_size):
    indices = np.arange(X.shape[0])
    np.random.shuffle(indices)
    for start_idx in range(0, X.shape[0], batch_size):
        batch_indices = indices[start_idx:start_idx + batch_size]
        yield X[batch_indices], y[batch_indices]

# Параметры SGD
lambda_reg = 0.001  # Регуляризационный коэффициент
batch_size = 10     # Размер батча
epochs = 10         # Количество эпох
learning_rate = 0.1 # Скорость обучения

# Инициализация весов
w = np.zeros(X.shape[1])

# Основной цикл обучения
for epoch in range(epochs):
    for X_batch, y_batch in batch_generator(X, y, batch_size):
        _, grad = logistic_loss_and_gradient(w, X_batch, y_batch, lambda_reg)
        w -= learning_rate * grad

# Вывод окончательных весов
print("Final weights:", w)
```

Если вы вводите новые переменные или делаете замены переменных, объясните, как восстановить оптимальные скорости из решения вашей задачи. Если выпуклость целевой функции или любой функции ограничения в вашей формулировке неочевидна, объясните, почему она выпуклая.

1. **Целевая функция**: Мы должны минимизировать суммарный расход топлива, который выражается через функцию $\(\Phi(s_i) = a s_i^2 + b s_i + c\)$. Поскольку $\(a = 1\)$, $\(b = 6\)$, и $\(c = 10\)$, целевая функция имеет вид:
   $\[ \text{minimize} \sum_{i=1}^n (s_i^2 + 6s_i + 10) \times \frac{d_i}{s_i} \]$
   Это функция выпуклая, так как $\(s_i^2 + 6s_i + 10\)$ является квадратичной и выпуклой функцией по $\(s_i\)$, а $\(d_i/s_i\)$ (время прохождения участка) является линейной функцией по $\(d_i\)$ и убывающей функцией по $\(s_i\)$. Произведение выпуклой возрастающей и убывающей функций также выпукло.

2. **Ограничения на скорость**: Для каждого участка $\(i\)$ скорость $\(s_i\)$ должна находиться в пределах заданных границ:
   $\[ s_i^\text{min} \leq s_i \leq s_i^\text{max} \]$

3. **Ограничения времени прохождения путевых точек**: Время прибытия на путевую точку $\(i\) (\(\tau_i\))$ можно выразить через скорости на участках:
   $\[ \tau_i = \sum_{j=1}^i \frac{d_j}{s_j} \]$
   Следовательно, ограничения времени принимают вид:
   $\[ \tau_i^\text{min} \leq \sum_{j=1}^i \frac{d_j}{s_j} \leq \tau_i^\text{max} \]$

Таким образом, задача выпуклой оптимизации формулируется следующим образом:

- $\( s_i^\text{min} \leq s_i \leq s_i^\text{max} \)$ для всех $\( i \)$

Пусть дан двунаправленный граф с множеством вершин $\mathcal{V}$ ($\mathcal{V} = \{1, \ldots, n \}$) и множеством дуг (направленных ребер) $\mathcal{E}$ ($\mathcal{E} = \{1, \ldots, m \}$). __а). (5 баллов)__ Объясните, какой физический смысл имеют переменные $f_{e}^{ij}$, и докажите, что формулировка задачи корректна. 

Переменные $f_e^{i j}$ в задаче о многопродуктовом потоке минимальной стоимости имеют следующий физический смысл:
- $f_e^{i j}$ представляет собой величину потока (например, объем груза, количество данных, количество транспортных средств и т.д.) который пересылается по дуге $e$ из вершины $i$ в вершину $j$.
- В контексте задачи, где вершины представляют города, а дуги - дороги между ними, $f_e^{i j}$ указывает на объем груза, который нужно перевезти по дороге $e$ из города $i$ в город $j$.

Теперь докажем, что формулировка задачи корректна:
1. **Неотрицательность потоков** $f_e^{i j} \geq 0$ : Это логично, поскольку поток не может быть отрицательным. В реальном мире вы не можете отправить отрицательное количество груза или данных.
2. **Ограничение пропускной способности** $\sum_{i, j \in \mathcal{V}} f_e^{i j} \leq b_e$ : Каждая дуга (или дорога) имеет максимальную пропускную способность, которую нельзя превышать. Это ограничение гарантирует, что общий поток по любой дуге не превысит ее пропускную способность.
3. **Баланс потоков**: Это условие обеспечивает, что для каждой пары вершин (городов) $i$ и $j$, суммарный поток из вершины $i$ равен количеству груза, который нужно отправить из $i$ в $j$ (т.е. $-d_{i j}$ ), суммарный поток в вершину $j$ равен этому же количеству (но с противоположным знаком, т.е. $d_{i j}$ ), и для всех остальных вершин суммарный поток равен нулю (то есть через любую другую вершину нет чистого входящего или исходящего потока для пары $i, j$ ). Это гарантирует, что весь трафик между парой городов будет корректно распределен по сети.
4. **Цель минимизации суммарной стоимости**: Это соответствует цели минимизировать общие затраты на перевозку по всем дугам сети, что является логичной целью в контексте оптимизации логистических или транспортных потоков.

__в). $\triangle$ (10 баллов)__ Будем считать, что в качестве решения нас интересуют только суммарный поток на каждой дуге. Переформулируйте задачу, уменьшив число переменных. Решите задачу в новой постановке с помощью `cvxpy`. Постройте график оптимального суммарного потока по дуге от индекса дуги, выведите оптимальную суммарную стоимость потока.

Чтобы уменьшить количество переменных и упростить задачу, мы можем сосредоточиться только на суммарном потоке через каждую дугу, а не на потоках для каждой пары вершин. Для этого мы создадим переменную $f_e$ для каждой дуги $e$, которая будет представлять суммарный поток через эту дугу. Это значительно уменьшит размерность задачи.

Переформулированная задача будет выглядеть следующим образом:
$$
\begin{array}{ll}
& \min _{f_e \in \mathbb{R}}\left[\sum_{e \in \mathcal{E}} c_e f_e\right] \\
\text { s.t. } & f_e \geq 0, \quad \forall e \in \mathcal{E} \\
& f_e \leq b_e \quad \forall e \in \mathcal{E} \\
& \sum_{e \in \mathcal{E}} A_{v e} f_e=\sum_{j \in \mathcal{V}} d_{v j}-\sum_{i \in \mathcal{V}} d_{i v}, \quad \forall v \in \mathcal{V}
\end{array}
$$

```python
# Переменная для суммарного потока по каждой дуге
f_e = cp.Variable(edges)

# Целевая функция - минимизация общей стоимости потоков
objective = cp.Minimize(cp.sum(cp.multiply(costs, f_e)))

# Ограничения
constraints = []

# Ограничение на неотрицательность и пропускную способность потоков
constraints += [f_e >= 0, f_e <= capacities]

# Ограничение баланса потоков для каждой вершины
for v in range(nodes):
    in_flow = cp.sum([f_e[e] for e in range(edges) if incidence_matrix[v, e] == -1])
    out_flow = cp.sum([f_e[e] for e in range(edges) if incidence_matrix[v, e] == 1])
    total_flow = cp.sum(traffic_matrix[v, :]) - cp.sum(traffic_matrix[:, v])
    constraints += [in_flow - out_flow == total_flow]

# Решение задачи оптимизации
problem = cp.Problem(objective, constraints)
optimal_cost = problem.solve(solver=cp.ECOS)

# Вывод оптимальной суммарной стоимости потока
print(f"Оптимальная суммарная стоимость потока: {optimal_cost}")

# График оптимального суммарного потока по дуге
plt.plot(f_e.value)
plt.xlabel('Индекс дуги')
plt.ylabel('Оптимальный суммарный поток')
plt.title('Оптимальный суммарный поток по дуге')
plt.show()
```

__г). $\triangle$ (10 баллов)__ Теперь вы можете улучшить сеть, увеличив пропускные способности дуг на произвольную величину $d > 0$, но суммарно не больше чем на $b_{add}$: $\sum_{e \in \mathcal{E}} d_e \leq b_{add}$. Также хочется учесть, что может произойти ЧС на одной из дуг: на любой (но только одной) дуге может произойти авария, что выразится в уменьшении её пропускной способности в два раза. 

Как адаптировать задачу под такую постановку? Найдите такой вектор $d$, который минимизирует стоимость оптимального потока в худшем случае (то есть минимизует максимальную по всем возможным авариям стоимость оптимального потока). Постройте график $d$ от индекса дуги, выведите стоимость оптимального потока в худшем случае.

Адаптация задачи под постановку:

1. **Переменная увеличения пропускной способности**: Введем вектор $d$ размерности 'edges', где каждый элемент $d_e$ обозначает увеличение пропускной способности для дуги $e$.
2. **Ограничение на суммарное увеличение**: Сумма всех $d_e$ должна быть не больше $b_{\text {add }}$.
3. **Наихудший сценарий (ЧС)**: Для каждой дуги е необходимо рассмотреть сценарий, когда ее пропускная способность уменьшается вдвое. Это означает решение задачи оптимизации потока для каждого такого сценария и выбор наихудшего (с максимальной стоимостью).
4. **Целевая функция**: Минимизация максимальной стоимости оптимального потока среди всех возможных аварий.

```python
# Количество вершин и рёбер
nodes, edges = incidence_matrix.shape

# Параметрический подход: различные значения для b_add
b_add_values = np.linspace(0, np.sum(capacities), 10)  # Например, от 0 до суммы всех capacities, разбитые на 10 шагов

# Сохранение результатов для каждого значения b_add
optimal_costs = []

for b_add in b_add_values:
    # Переменные для увеличения пропускной способности и потоков
    d = cp.Variable(edges)
    f_e = cp.Variable(edges)

    # Ограничения
    constraints = [f_e >= 0, f_e <= capacities + d, cp.sum(d) <= b_add]

    # Ограничение баланса потоков для каждой вершины
    for v in range(nodes):
        in_flow = cp.sum([f_e[ee] for ee in range(edges) if incidence_matrix[v, ee] == -1])
        out_flow = cp.sum([f_e[ee] for ee in range(edges) if incidence_matrix[v, ee] == 1])
        total_flow = cp.sum(traffic_matrix[v, :]) - cp.sum(traffic_matrix[:, v])
        constraints += [in_flow - out_flow == total_flow]

    # Целевая функция - минимизация стоимости потока
    objective = cp.Minimize(cp.sum(cp.multiply(costs, f_e)))

    # Решение задачи оптимизации
    problem = cp.Problem(objective, constraints)
    optimal_cost = problem.solve(solver=cp.ECOS)

    # Сохранение оптимальной стоимости
    optimal_costs.append(optimal_cost)

# График зависимости оптимальной стоимости от b_add
plt.plot(b_add_values, optimal_costs)
plt.xlabel('b_add')
plt.ylabel('Оптимальная стоимость потока')
plt.title('Зависимость оптимальной стоимости от увеличения пропускной способности')
plt.show()
```

```python
# Устанавливаем значение b_add
b_add = 0.10 * np.sum(capacities)  # 10% от суммарной пропускной способности

# Создаем переменную для увеличения пропускной способности
d = cp.Variable(edges, nonneg=True)

# Создаем переменные для потоков, учитывая аварию на каждой дуге
f_worst_case = cp.Variable((edges, edges), nonneg=True)

# Создаем выражения стоимости для всех сценариев
worst_case_costs = []

# Ограничения
constraints = [cp.sum(d) <= b_add, d <= capacities]

# Добавляем ограничения для каждого сценария аварии
for e in range(edges):
    # Пропускная способность для сценария аварии на дуге 'e'
    scenario_capacities = np.copy(capacities)
    scenario_capacities[e] /= 2

    # Ограничения пропускной способности с учетом возможной аварии и улучшений
    constraints += [f_worst_case[e, :] <= scenario_capacities + d]

    # Стоимость для сценария 'e'
    cost_e = cp.sum(cp.multiply(costs, f_worst_case[e, :]))
    worst_case_costs.append(cost_e)

    # Ограничения баланса потоков для каждой вершины
    for v in range(nodes):
        constraints += [
            cp.sum(cp.multiply(incidence_matrix[v, :], f_worst_case[e, :])) ==
            cp.sum(traffic_matrix[v, :]) - cp.sum(traffic_matrix[:, v])
        ]

# Используем cp.hstack для создания одного выражения из списка worst_case_costs
max_cost = cp.max(cp.hstack(worst_case_costs))

# Задача оптимизации
problem = cp.Problem(cp.Minimize(max_cost), constraints)

# Решение задачи
problem.solve(solver=cp.ECOS)

# Результаты
print(f"Увеличение пропускной способности для каждой дуги: {d.value}")
print(f"Максимальная стоимость в худшем случае: {max_cost.value}")

# Построение графика
plt.figure(figsize=(12, 6))
plt.bar(range(edges), d.value)
plt.title('Увеличение пропускной способности для каждой дуги')
plt.xlabel('Индекс дуги')
plt.ylabel('Увеличение пропускной способности d_e')
plt.show()
```
