Заметим, что первый пункт не является блокирующим, так как можно использовать готовые библиотечные реализации.

### Реализация критериев (0.5 баллов)

```python
import numpy as np
import math

def pearson(x, y):
    # Вычислим средние значения x и y
    mean_x = np.mean(x)
    mean_y = np.mean(y)

    # Вычисление ковариации и стандартных отклонений
    covariance = np.mean((x - mean_x) * (y - mean_y))
    std_x = np.std(x, ddof=0)
    std_y = np.std(y, ddof=0)

    # Вычисление коэффициента корелляции Пирсона
    return covariance / (std_x * std_y)



def spearman(x, y):
    # Ранжируем данные
    rank_x = np.argsort(np.argsort(x))
    rank_y = np.argsort(np.argsort(y))

    # Теперь вызовем функцию Пирсон на ранжированных данных
    return pearson(rank_x, rank_y)
```

```python
import scipy.stats as sps

for i in range(100):
    x = sps.norm.rvs(size=100)
    y = sps.norm.rvs(size=100)

    assert np.isclose(sps.pearsonr(x,y)[0], pearson(x,y))
    assert np.isclose(sps.spearmanr(x,y)[0], spearman(x,y))
```

```python
import matplotlib.pyplot as plt
import scipy.stats as sps
import numpy as np

n, m = 3, 3
data = [[None] * m for _ in range(n)]
size = 250

# 1. Идеальная линейная зависимость
sample_0 = sps.norm.rvs(size=size)
data[0][0] = (sample_0, 2 * sample_0)

sample_1 = sps.expon.rvs(size=size)

# 2. Линейная зависимость с шумом
data[0][1] = (sample_0, 2 * sample_0 + np.random.normal(0, 1, size))

# 3. Нелинейная зависимость (квадратичная)
data[0][2] = (sample_0, sample_0**2)

# 4. Случайные данные (независимые переменные)
data[1][0] = (sample_0, np.random.normal(0, 1, size))

# 5. Монотонная, но нелинейная зависимость (логарифмическая)
data[1][1] = (sample_1, np.log(sample_1 + 0.01))  # +0.1 чтобы избежать log(0)

# 6. Переменные с разными масштабами
data[1][2] = (sample_0, 100 * sample_0)

# 7. Переменные с обратной линейной зависимостью
data[2][0] = (sample_0, -sample_0)

# 8. Данные с кластерами
cluster_1 = np.random.normal(-5, 1, size // 2)
cluster_2 = np.random.normal(5, 1, size // 2)
data[2][1] = (np.concatenate([cluster_1, cluster_2]), np.concatenate([cluster_1, -cluster_2]))

# 9. Синусоидальная зависимость
data[2][2] = (sample_0, np.sin(sample_0))

fig, axs = plt.subplots(n, m, figsize=(12,12))

fig.suptitle('Pearson Correlation Examples')

for i in range(n):
    for j in range(m):
        ax = axs[i][j]
        x, y = data[i][j]
        ax.set_title(f'{sps.pearsonr(x, y)[0]:.2f}')
        ax.scatter(x, y, color='b', alpha=1)
        ax.xaxis.set_visible(False)
        ax.yaxis.set_visible(False)

plt.show()
```

4. Случайные данные: Оба коэффициента корреляции показывают низкую корреляцию, что соответствует независимости данных.

5. Логарифмическая зависимость: Пирсон ловит неплохо, выдавая 0.8, но это все-таки не 0.95, а Спирмен отрабатывает идеально и выдает чистый 1.00.

6. Разные масштабы: оба коэффициента вообще никак не чувствуют домножение всей выборки на какую-то положительную константу, в силу того, что их формулы просто однородны (у Пирсона просто очевидно из всего выносится эта константа, а у Спирмена - ранги не меняются при домножении на константу), поэтому выдается то же что и в 1 - чистый 1.00

7. Обратная линейная зависимость: аналогично прошлому пункту, просто знак поменялся из-за знака константы.

8. Кластеризованные данные: Пирсон может быть введен в заблуждение кластерами, не показывая явной зависимости. Спирмен, оценивающий монотонность, также может не показать чёткой корреляции. Данные и кусочно-линейны, и кусочно-монотонны, но все равно не получается - как в пункте 3, только еще хуже.

9. Синусоидальная зависимость: оба коэффициента выдали довольно близкий к 1 результат (порядка 0.9), он не космически близок к единице, так как синусоида не совсем монотонна (но весьма близка, то есть, данные почти не шаффлятся синусом, поэтому коэффициент хорошо отрабатывает), ситуация сильно лучше, чем в пункте 3, ибо там были две равноправные ветви параболы, а тут маленькие хвостики (вспомним, что мы семплировали данные нормально, поэтому в хвостики синусоиды точки попадают крайне редко, в итоге имеем огромную корреляции в силу монотонности на множестве очень большой вероятности).
